## Linear Regression On Menlo Therapeutics Stock Prices

### Data Preprocessing

### Import all necessary libraries, module and function

```python
# import pandas library to read .csv / .xlsx file
#import numpy to suppress the numbers when .values is being used
# Getting the library scikitlearn, the module model_selection and the function train_test_split for training and testing the
# X and y variables

# import matplotlib for plotting
# to display large numbers in plotting
# from scikitlearn library, get the linear_model module to import LinearRegression class to build and train the model

import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker 
from sklearn.linear_model import LinearRegression 
```

##### Importing The Dataset

```python
# use pandas to read the dataset

dataset = pd.read_csv('Menlo Therapeutics Stock.csv')
```

```python
#display the data

print("Menlo Therapeutics - Stock Price (January 2018 - April 2020)")
dataset.head()
```

```python
#displaying the info of data

print("Overview of the features and their types:")
dataset.info()

# if the number of non-null is equal to the number of entries means there are no null cell
```

##### Importing The Clean Dataset

```python
# importing the clean dataset

clean_dataset = pd.read_csv('Menlo Therapeutics Stock - Cleaned.csv')
```

```python
#display the cleaned dataset 
clean_dataset.head()
```

##### Extracting The Inputs And Output

```python
# Getting the independent variables "X"

X = clean_dataset.iloc[:,1:-1].values
np.set_printoptions(suppress=True) # suppress makes the integers display as whole number instead of in scientific form
X 
```

```python
# Getting the Dependent Variables "y"

y = clean_dataset.iloc[:,-1].values
y
```

##### Creating the Training Set and the Test Set

```python
# Getting the library scikitlearn, the module model_selection and the function train_test_split for training and testing the X and y variables

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 440 )
```

```python
# display trained X values

X_train
```

```python
# display X variables for test

X_test
```

```python
# display trained y variables

y_train
```

```python
#display y values for test

y_test
```

### Plotting the linearity relation of each independent variable to the dependent variable with trend line

```python
X_index1 = X[:,0] # X0 index
X_index2 = X[:,1] # X1 index
X_index3 = X[:,2] # X2 index
X_index4 = X[:,3] # X3 index

# Plottting into 1 image

fig, axs = plt.subplots(2, 2, figsize=(10, 4))

# plot 1
m, b = np.polyfit(X_index1,y, 1)
axs[0, 0].scatter(X_index1, y, color='red', label='Openning Value', s = 20)  
axs[0, 0].set_xlabel('X1 (Openning Value)')
axs[0, 0].set_ylabel('y (Closed Value)')
axs[0, 0].set_title('Partial Plot: X1 vs y')
axs[0, 0].plot(X_index1, X_index1 * m + b, color='salmon', label='Trend Line') 
axs[0, 0].grid()

# plot 2
m, b = np.polyfit(X_index2,y, 1)
axs[0, 1].scatter(X_index2, y, color='yellow', label='Highest Value', s = 20)  
axs[0, 1].set_xlabel('X2 (Highest Value)')
axs[0, 1].set_ylabel('y (Closed Value)')
axs[0, 1].set_title('Partial Plot: X2 vs y')
axs[0, 1].plot(X_index2, X_index2 * m + b, color='salmon', label='Trend Line') 
axs[0, 1].grid()

# plot 3
m, b = np.polyfit(X_index3,y, 1)
axs[1, 0].scatter(X_index3, y, color='green', label='Lowest Value', s = 20)  
axs[1, 0].set_xlabel('X3 (Lowest Value)')
axs[1, 0].set_ylabel('y (Closed Value)')
axs[1, 0].set_title('Partial Plot: X3 vs y')
axs[1, 0].plot(X_index3, X_index3 * m + b, color='salmon', label='Trend Line') 
axs[1, 0].grid()

# plot 4
m, b = np.polyfit(X_index4,y, 1)
axs[1, 1].scatter(X_index4, y, color='blue', label=' Volume', s = 20)  
axs[1, 1].set_xlabel('X4 (Volume)')
axs[1, 1].set_ylabel('y (Closed Value)')
axs[1, 1].set_title('Partial Plot: X4 vs y')
axs[1, 1].plot(X_index4, X_index4 * m + b, color='salmon', label='Trend Line') 
plt.gca().xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
axs[1, 1].grid()

# Adjust layout
plt.tight_layout()

# Display the combined plots
plt.show()
```

## Building and Training the Model

### Building the model

```python
# from scikitlearn library, get the linear_model module to import LinearRegression class

model = LinearRegression()
```

##### Training the Model

```python
# Train the model using the variable for training (X_train, y_train) by using the .fit method inside the class LinearRegression

model.fit(X_train, y_train)
```

##### Inference

```python
# predicting the y variable using model and the test X variables 

y_pred = model.predict(X_test)
y_pred
```

```python
y_test

# comparing the predicted and test values
```

##### Plotting the Actual vs Predicted values

```python
# plotting the predicted and tested values

X_test_index = X_test[:,2]

plt.figure(figsize=(10, 4))
plt.scatter(X_test_index, y_test,  label='Actual Values', s = 20)  # Actual test values
plt.scatter(X_test_index, y_pred,  label='Predicted Values',  s = 20)  # Predicted values

# plotting the Ideal Regression Line
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='black', linestyle='-', label='Ideal Prediction Line')


# Add labels and legend
plt.xlabel('X1 (Openning Value)')
plt.ylabel('y (Dependent Variable)')
plt.title('Partial Plot: X1 vs Actual and Predicted')
plt.legend()
plt.grid(True)
# Show the plot
plt.show()
```

#### Making the prediction of a single data point with open = 35.6523 ,high =37.8547 , low = 32.00247, volume = 925696

```python
# predict the output using the linear regression model with the parameters above

predict = model.predict([[20.5,31.1,20.5,5253400]])

print(f'The predicted value for this parameters is {predict}')
```

## Evaluating the Model

### R-Squared

```python
# determine the R_Squared for evaluating the model

from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(f'R-Squared: {r2}')

# based on the result of R-Squared, this model is very good or suspicious
```

### Adjusted R-Squared

```python
k = X_test.shape[1]
k   
```

```python
n =  X_test.shape[0]
n
```

```python
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
print(f'Adjusted R-Squared: {adj_r2}')

# based on the result of Adjusted R-Squared, this model is very good or suspicious
```

### Mean Squared Error

```python
# from scikitlearn we import the module mean_squared_error to find the Mean Squared Error

from sklearn.metrics import mean_squared_error 
MSE = mean_squared_error(y_test, y_pred)
MSE

print(f'Mean Squared Error: {MSE}')
```

### Adding Noise 

```python
# add noise to reduce R-Squared

y_noisy = y + np.random.randn(550) * 1.5 # noise level
y_noisy
```

### Creating a new training set for the noisy variable

```python
X_train_noisy, X_test_noisy, y_train_noisy, y_test_noisy = train_test_split(X, y_noisy, test_size = .2, random_state = 440)
```

### Fit the noisy dependent variable into a new model

```python
# Fit a new model

model.fit(X_train_noisy, y_train_noisy)
```

```python
y_test_noisy
```

```python
y_noisy_pred = model.predict(X_test_noisy)
y_noisy_pred
```

```python
X_train_noisy
```

### Plot The new noisy Predicted Values

```python
# plotting the predicted and tested values

X_test_index = X_test_noisy[:,2]

plt.figure(figsize=(10,4))
plt.scatter(X_test_index, y_test_noisy, label='Actual Values', s = 20)  # Noised Actual test values
plt.scatter(X_test_index, y_noisy_pred,  label='Predicted Values',  s = 20)  # Predicted values

# plotting the Ideal Regression Line
plt.plot([y_test_noisy.min(), y_test_noisy.max()], [y_test_noisy.min(), y_test_noisy.max()], color='black', linestyle='-', label='Ideal Prediction Line')


# Add labels and legend
plt.xlabel('X1 (Openning Value)')
plt.ylabel('y (Dependent Variable)')
plt.title('Partial Plot: X1 vs Noised Actual and Predicted')
plt.legend()
plt.grid(True)
# Show the plot
plt.show()
```

### New R-Squared

```python
# Calculate new R-Squared
new_r2 = r2_score(y_test_noisy, y_noisy_pred)
print("New R-Squared:", new_r2)

# results varies due to the randomness of the noise introduced
```

### New Means Squared Error

```python
from sklearn.metrics import mean_squared_error 
MSE = mean_squared_error(y_test_noisy, y_noisy_pred)
MSE

print(f'New Mean Squared Error: {MSE}')

# results varies due to the randomness of the noise introduced
```
