## Linear Regression On Menlo Therapeutics Stock Prices

### Data Preprocessing

##### Importing The Dataset

```python
# import pandas to read .csv / .xlsx file
# use pandas to read the dataset

import pandas as pd
dataset = pd.read_csv('Menlo Therapeutics Stock.csv')
```

```python
#display the data

print("Menlo Therapeutics - Stock Price (January 2018 - April 2020)")
dataset.head()
```

```python
#displaying the info of data

print("Overview of the features and their types:")
dataset.info()

# if the number of non-null is equal to the number of entries means there are no null cell
```

##### Importing The Clean Dataset

```python
# importing the clean dataset

clean_dataset = pd.read_csv('Menlo Therapeutics Stock - Cleaned.csv')
```

```python
#display the cleaned dataset 
clean_dataset.head()
```

##### Extracting The Inputs And Output

```python
# Getting the independent variables "X"
#import numpy to suppress the numbers when .values is being used
# note: it does not affect the final output, just the displayed value

import numpy as np
X = clean_dataset.iloc[:,1:-1].values
np.set_printoptions(suppress=True)
X 
```

```python
# Getting the Dependent Variables "y"

y = clean_dataset.iloc[:,-1].values
y
```

##### Creating the Training Set and the Test Set

```python
# Getting the library scikitlearn, the module model_selection and the function train_test_split for training and testing the X and y variables

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state =0 )
```

```python
# display trained X values

X_train
```

```python
# display X variables for test

X_test
```

```python
# display trained y variables

y_train
```

```python
#display y values for test

y_test
```

## Building and Training the Model

### Building the model

```python
# linear_model is the module
# `LinearRegression is a class` is defining that `LinearRegression` is a class within the `linear_model` module. It indicates that `LinearRegression` is a blueprint or template for creating objects that represent linear regression models.
# Class is a pre-coded blueprint of something we want to build from which objects are created.

# from scikitlearn library, get the linear_model module to import LinearRegression class

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

##### Training the Model

```python
# fit is a method inside LinearRegression class - they are like functions.
# Train the model using the variable for training (X_train, y_train) by using the .fit method inside the class LinearRegression

model.fit(X_train, y_train)
```

##### Inference

```python
# predicting the y variable using model and the test X variables 

y_pred = model.predict(X_test)
y_pred
```

```python
y_test

# comparing the predicted and test values
```

##### Plotting the Actual vs Predicted values

```python
# plotting the predicted and tested values
# import matplotlib
import matplotlib.pyplot as plt


X_test_index = X_test[:,2]

plt.figure(figsize=(12, 8))
plt.scatter(X_test_index, y_test, color='red', label='Actual Values', s = 20)  # Actual test values
plt.scatter(X_test_index, y_pred, color='blue', label='Predicted Values', marker='x', s = 20)  # Predicted values

# plotting the Ideal Regression Line
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='black', linestyle='-', label='Ideal Prediction Line')


# Add labels and legend
plt.xlabel('X1 (First Independent Variable)')
plt.ylabel('y (Dependent Variable)')
plt.title('Partial Plot: X1 vs Actual and Predicted')
plt.legend()

# Show the plot
plt.show()
```

#### Making the prediction of a single data point with open = 35.6523 ,high =37.8547 , low = 32.00247, volume = 925696

```python
# predict the output using the linear regression model with the parameters above

predict = model.predict([[35.6523,37.8547,32.00247,925696]])

print(f'The predicted value for this parameters is {predict}')
```

## Evaluating the Model

### R-Squared

```python
# determine the R_Squared for evaluating the model

from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(f'R-Squared: {r2}')

# based on the result of R-Squared, this model is very good
```

### Adjusted R-Squared

```python
k = X_test.shape[1]
k   
```

```python
n =  X_test.shape[0]
n
```

```python
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
print(f'Adjusted R-Squared: {adj_r2}')

# based on the result of Adjusted R-Squared, this model is very good
```

### Mean Squared Error

```python
# from scikitlearn we import the module mean_squared_error to find the Mean Squared Error

from sklearn.metrics import mean_squared_error 
MSE = mean_squared_error(y_test, y_pred)
MSE

print(f'Mean Squared Error: {MSE}')
```
